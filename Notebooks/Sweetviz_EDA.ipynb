EDA using the pandas and import dtale libraries:

```python
import pandas as pd
import sweetviz
%matplotlib inline
```

2. Load the .csv into a dataframe using read_csv:

```python
housing_data = pd.read_csv("/content/Melbourne_housing_EDA.csv")
```

3. Create an Automated EDA report using the analyze method in the sweetviz library. Use the show_html method to export the report to an HTML file:

```python
viz_report = sweetviz.analyze(housing_data, target_feat = 'Price')
```

```python
viz_report.show_html()
```
